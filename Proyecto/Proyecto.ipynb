## Introducción

Es de ínteres del ingeniero mecánico idear nuevas formas de reducir las pérdidas en sistemas de conversión, transporte y transformación de energía. Esto se logra produciendo dispositivos que aprovechen la energía residual del sistema, ya sea en forma de calor o de vibraciones mecánicas, y generen otra forma de energía. Esta área de la ingeniería es llamada harvesting. Para aprovechar las energía que brinda las vibraciones se ha diseñado un sistema que consiste en una lámina biestable y en un perfil aerodinámico. 

## Objetivos

El objetivo general del proyecto es caracterizar el sistema en sus dos regímenes: lineal y caótico. Para ello debemos cumplir los siguientes objetivos específicos:

## Marco Teórico

###Lámina biestable
En el proceso de producción de la lámina, que es compuesta por un apilamiento de segmentos de fibra de carbono unidireccional, se generan esfuerzos residuales debido a la contracción y dilatación de la lámina en direcciones perpendiculares sobre el plano xy. Esto genera la biestabilidad de la lámina.

La biestabilidad brinda varias características a este sistema, sin embargo, la posibilidad de tener distintos comportamientos al estar excitada por un movimiento oscilatorio con cierta frecuencia es la característica por la que se usa en el área de harvesting. Cuando la lámina esta sujeta a un movimiento oscilatorio, existen 2 respuestas de la lámina biestable:


- Al estar excitada por una frecuencia igual a la frecuencia de resonancia de alguno de los estados estables, pasa de un estado estable a otro y se quede en el segundo estado, esto es llamado snap-through.

- Al estar excitada por distintas frecuencias dentro de un rango, pase de un estado estable a otro aperiódicamente, esto es llamado régimen caótico.

###Perfil aerodinámico
En la universidad se encontró que la aceleración del movimiento armónico debido a las vibraciones de sistemas mecánicos no es suficiente para generar el régimen caótico en las láminas producidas, es por eso que se seleccionó un perfil aerodinámico. El perfil aerodinámico fue usado para idear otra forma de excitar el sistema a partir de cargas aerodinámicas. Se caracteriza por tener geometrias en su superficie superior e inferior que producen una diferencia de presión entre las superficies. Esta diferencia de presión da lugar a dos fuerzas importantes en el movimiento del perfil. La fuerza de sustentación es la fuerza vertical debido a la diferencia de presiones y la fuerza de arrastre es la fuerza horizontal debido a la diferencia de presiones. Además las fuerzas dependen del ángulo entre el flujo y el perfil (ángulo de ataque), del area del perfil y el largo del perfil. Es por esto que es común encontrar graficas del desempeño de un perfil en función del ángulo de ataque.

###Lamina biestable como sistema no lineal
Un sistema no lineal se puede describir a partir de una matriz 2$\times$2, al expresar el movimiento con las variables de posición y velocidad, de la siguiente manera:

$A = \left( \begin{array}{ccc}
a & b \\
c & d \\ \end{array} \right)$

El polinomio característico del sistema se encuentra de calcular $det(A - \lambda I) = 0$, donde I es la matriz identidad. Las raíces o puntos de equilibrio resultantes de este polinomio son:

$\lambda_{1} = \frac {\tau+\sqrt{\tau^2-4\Delta}}{2}$
$\lambda_{2} = \frac {\tau-\sqrt{\tau^2-4\Delta}}{2}$

Donde $\Delta$ y $\tau$ son el determinante y la traza de la matriz $A$, respectivamente.

Sí las raíces del polinomio son reales, los puntos de equilibrio del sistema son llamados nodos, que pueden ser puntos silla, puntos estables o puntos inestables. Sí las raíces del polinomio son complejas, los puntos de equilibrio del sistema son llamados centros o espirales. Los centros son neutralmente estables, mientras que las espirales pueden ser estables o inestables. 

![](https://raw.githubusercontent.com/deolaya1318/MCdata/master/puntos.png)Clasificación de los puntos de equilibrio en función del determinante y la traza de la matriz A. Tomado de: Strogatz (1994).

Adicionalmente, podemos observar el tipo de punto de equilibrio al graficar en el espacio de fases la variable de estado y su derivada.

![](https://raw.githubusercontent.com/deolaya1318/MCdata/master/fases.png)Espacio de fases de puntos de equilibrio típicos de sistemas no lineales. Tomado de: Strogatz (1994).

Las orbitas son características de sistemas que tienen soluciones periódicas, como lo son los osciladores armónicos simples que oscilan con cierta frecuencia. Mientras que las espirales representan osciladores no lineales que al tener amortiguamiento pueden tener oscilaciones que decaen o aumentan exponencialmente. Estas corresponden a espirales estables e inestables, respectivamente. Estos tipos de puntos de equilibrio son observables experimentalmente en sistemas modelados por la ecuación de Duffing.

Es por esto que se ha modelado el sistema a partir de un balance de fuerzas en donde la fuerza que deflecta la viga en voladizo, es decir la lámina biestable, se explica a partir de la ecuación del resorte de Duffing, que consiste en un sistema no lineal de masa-resorte:

$m\ddot{z} + c\dot{z} + k_1z + k_2z^2 + k_3z^3 = F_s$

Con la ayuda de la siguiente figura podemos entender esta ecuación. 

![Diagrama del modelo físico de la lámina biestable. Tomado de: Monsalve (2014)](https://raw.githubusercontent.com/deolaya1318/MCdata/master/duffing.png)Diagrama del modelo físico de la lámina biestable. Tomado de: Monsalve (2014).

La variable que designa la fuerza de sustentación debido al perfil es $F_s$ y la variable que designa la deflexión de la viga es $z$. Adicional al término lineal debido al resorte existen dos términos adicionales. El término cúbico de la ecuación, origina un potencial de doble pozo en el sistema de viga en voladizo. Evidencia experimental ha mostrado que existe un comportamiento asimétrico de la lámina biestable, en donde uno de los estados es más estable que el otro. Es por esto que se incluye un término cuadrático a la ecuación. Las constantes $m$, $c$, $k_1$, $k_2$ y $k_3$ dan cuenta de las propiedades de inercia, disipación de la energía y restauración de la energía del sistema biestable.

En la primera gráfica podemos notar que a esa velocidad del viento, el sistema tiene oscilaciones con una amplitud y un periodo definido. Decimos que se comporta como un oscilador lineal alrededor del primer estado estable. A partir de una velocidad de 12.2 m/s, podemos notar oscilaciones débiles alrededor de ambos estados estables que aumentan su amplitud en función de la velocidad del viento. El paso entre un estado y otro no tiene un periodo definido, lo que caracteriza al régimen caótico de las láminas. En cada una de las gráficas se notan lineas entrecortadas que representan el inicio y el fin de las ventanas seleccionadas para realizar un análisis posterior de Fourier. 

```python
#Ahora queremos graficar el espacio de fases de la posición angular, para ello debemos hallar la velocidad angular
#del dispositivo. Usamos el método de diferencias centrales para encontrar el valor de la derivada de la posición
#angular. Expresamos esta velocidad en unidades de rad/s, por eso debemos incluir un factor de conversión dentro
#de la función.
def derivate(data):
    h = data[1,0] - data[0,0]
    dataforward = np.roll(data[:,1],-1)
    databackward = np.roll(data[:,1],1)
    conv = np.pi/180
    omega = conv*(dataforward-databackward)/(2.*h)
    return omega[1:-1]
#Aplicamos la función derivada a cada uno de los datos
omega1 = derivate(data1)
omega2 = derivate(data2)
omega3 = derivate(data3)
omega4 = derivate(data4)
omega5 = derivate(data5)

#Para facilidad al graficar la velocidad angular en función del tiempo se guardan los tiempos y los resultados de
#aplicar la función derivate en dos array
tiempo = np.array([data1[1:-1,0],data2[1:-1,0],data3[1:-1,0],data4[1:-1,0],data5[1:-1,0]])
omega = np.array([omega1,omega2,omega3,omega4,omega5])

#Usando un ciclo y condicionales graficamos los resultados en una grilla como fueron presentados en el paso anterior
fig = figure(figsize=(15,15))
fig.suptitle(u"Velocidad angular en función de la velocidad del viento - Diferencias centrales",fontsize=20)
for i in range(5):
    if i == 0 or i == 1:
        ax = plt.subplot2grid((3,2),(0,i))
        ax.plot(tiempo[i],omega[i])
        ax.set_xlim(0,max(tiempo[i]))
        ax.set_title('Velocidad del viento = '+str(velocity[i])+' m/s')
    elif i == 2 or i == 3:
        ax = plt.subplot2grid((3,2),(1,i-2))
        ax.plot(tiempo[i],omega[i])
        ax.set_xlim(0,max(tiempo[i]))
        ax.set_title('Velocidad del viento = '+str(velocity[i])+' m/s')
    else:
        ax = plt.subplot2grid((3,2),(2,0),colspan=2)
        ax.plot(tiempo[i],omega[i])
        ax.set_xlim(0,max(tiempo[i]))
        ax.set_title('Velocidad del viento = '+str(velocity[i])+' m/s')
fig.text(0.08,0.55,r'$\omega$ [rad/s]',rotation='vertical', fontsize=12)
fig.text(0.5,0.08,u'Tiempo [s]',rotation='horizontal', fontsize=12)
plt.show()
```

Aqui podemos corroborar que antes de llegar a una velocidad del viento de 12.2 m/s, la lámina biestable acoplada al perfil aerodinámico se comporta como un oscilador lineal, pues la amplitud y el periodo de la velocidad angular no cambian con respecto al tiempo. Mientras que para velocidades igual y mayor a esta, existe oscilaciones alrededor de los dos estados estables y entre ellos.

```python
#Ahora procedemos a graficar el espacio de fases de la posición angular. Nuevamente guardamos el ángulo en un array
#para facilitar el uso de ciclos en la gráfica realizada. Nuevamente hacemos uso de un ciclo y de tres expresiones
#con condicionales para graficar los resultados en una configuración subplot2grid.
angle = np.array([data1[1:-1,1],data2[1:-1,1],data3[1:-1,1],data4[1:-1,1],data5[1:-1,1]])
fig = figure(figsize=(15,15))
fig.suptitle(u"Espacio de fases del ángulo",fontsize=20)
for i in range(5):
    if i == 0 or i == 1:
        ax = plt.subplot2grid((3,2),(0,i))
        ax.plot(angle[i],omega[i])
        ax.set_title('Velocidad del viento = '+str(velocity[i])+' m/s')
    elif i == 2 or i == 3:
        ax = plt.subplot2grid((3,2),(1,i-2))
        ax.plot(angle[i],omega[i])
        ax.set_title('Velocidad del viento = '+str(velocity[i])+' m/s')
    else:
        ax = plt.subplot2grid((3,2),(2,0),colspan=2)
        ax.plot(angle[i],omega[i])
        ax.set_title('Velocidad del viento = '+str(velocity[i])+' m/s')
fig.text(0.08,0.55,r'$\omega$ [rad/s]',rotation='vertical',fontsize=12)
fig.text(0.5,0.08,r'$\theta$ [grados]',rotation='horizontal',fontsize=12)
plt.show()
```

Esta gráfica nos permite inferir el tipo de punto de equilibrio en función de la velocidad del viento. Para velocidades menores a 12.2 m/s, los puntos de equilibrio del sistema son centros y son puntos estables. A partir de esta velocidad, los puntos de equilibrio son espirales inestables, pues de acuerdo a estas tres graficas presentan oscilaciones alrededor de un punto estable que aumentan su amplitud hasta llegar a un punto donde la energía del sistema es suficiente para realizar el salto al otro punto estable. Cuando realiza el salto, el sistema tiene oscilaciones que aumentan su amplitud hasta llegar a una amplitud que lo hace volver al estado inicial. Esto ocurre continuamente y sin un periodo definido, hemos llegado al régimen caótico!

Sin embargo, al seguir aumentando la velocidad del viento, podemos notar que los puntos de equilibrio pueden ser espirales inestables o orbitas alrededor de ambos puntos de equilibrio. Esto quiere decir que seguir aumentando la velocidad nos hace regresar al regimen lineal donde ocurren oscilaciones con igual amplitud y periodo para todos los tiempos. 

```python
#En este paso usamos slice notation para hacer las ventanas de ínteres que se muestran en la primera grafica
window1 = data1[1600:2000]
window2a = data2[800:1000]
window2b = data2[1500:2000]
window3a = data3[830:1000]
window3b = data3[2000:3000]
window4a = data4[800:1300]
window4b = data4[2000:2500]
window4c = data4[3150:3200]
window5a = data5[1120:1190]
window5b = data5[1300:1600]
window5c = data5[1800:2400]
#Para graficar las ventanas guardamos los resultados del slice notation y de la derivada de la posición angular
#en arrays. Luego usamos un ciclo para graficar con puntos el espacio de fases de cada ventana para cada velocidad
#del viento
windows = np.array([window1[:,1],window2a[:,1],window2b[:,1],window3a[:,1],window3b[:,1],window4a[:,1],window4b[:,1]
                    ,window4c[:,1],window5a[:,1],window5b[:,1],window5c[:,1]])
omegawindows = np.array([omega1[1600:2000],omega2[800:1000],omega2[1500:2000],omega3[830:1000],omega3[2000:3000]
                        ,omega4[800:1300],omega4[2000:2500],omega4[3150:3200],omega5[1120:1190],omega5[1300:1600]
                        ,omega5[1800:2400]])
fig = figure(figsize=(15,15))
fig.suptitle(u"Espacio de fases en las ventanas a analizar",fontsize=20)
for i in range(5):
    if i == 0:
        ax = plt.subplot2grid((3,2),(0,i))
        ax.plot(windows[i],omegawindows[i],'go',label = 'ventana '+str(i+1))
        ax.set_title('Velocidad del viento = '+str(velocity[i])+' m/s')
        ax.legend()
    elif i == 1:
        ax = plt.subplot2grid((3,2),(0,i))
        ax.plot(windows[i],omegawindows[i],'go',label = 'ventana '+str(i))
        ax.plot(windows[i+1],omegawindows[i+1],'ro',label = 'ventana '+str(i+1))
        ax.set_title('Velocidad del viento = '+str(velocity[i])+' m/s')
        ax.legend()
    elif i == 2:
        ax = plt.subplot2grid((3,2),(1,i-2))
        ax.plot(windows[2*i-1],omegawindows[2*i-1],'go',label = 'ventana '+str(i-1))
        ax.plot(windows[2*i],omegawindows[2*i],'ro',label = 'ventana '+str(i))
        ax.set_title('Velocidad del viento = '+str(velocity[i])+' m/s')
        ax.legend()
    elif i == 3:
        ax = plt.subplot2grid((3,2),(1,i-2))
        ax.plot(windows[2*i-1],omegawindows[2*i-1],'go',label = 'ventana '+str(i-2))
        ax.plot(windows[2*i],omegawindows[2*i],'ro',label = 'ventana '+str(i-1))
        ax.plot(windows[2*i+1],omegawindows[2*i+1],'co',label = 'ventana '+str(i))
        ax.set_title('Velocidad del viento = '+str(velocity[i])+' m/s')
        ax.legend()
    else:
        ax = plt.subplot2grid((3,2),(2,0),colspan=2)
        ax.plot(windows[2*i],omegawindows[2*i],'go',label = 'ventana '+str(i-3))
        ax.plot(windows[2*i+1],omegawindows[2*i+1],'ro',label = 'ventana '+str(i-2))
        ax.plot(windows[2*i+2],omegawindows[2*i+2],'co',label = 'ventana '+str(i-1))
        ax.set_title('Velocidad del viento = '+str(velocity[i])+' m/s')
        ax.legend()
fig.text(0.08,0.55,r'$\omega$ [rad/s]',rotation='vertical',fontsize=12)
fig.text(0.5,0.08,r'$\theta$ [grados]',rotation='horizontal',fontsize=12)
plt.show()
```

Cuando la velocidad del viento es de 11.5 m/s, el espacio de fases de la ventana seleccionada muestra que existe un solo punto alrededor del cual oscila la lámina, representado por el centro de la orbita. Para las tres siguientes velocidades, se notan espirales alrededor de ambos puntos estables. Sin embargo, una de estas espirales es mucho más estable que la otra, pues existen una mayor de cantidad de orbitas que aumentan su tamaño alrededor del punto más estable. Cuando la velocidad del viento ha llegado a 13,9 m/s podemos notar espirales alrededor de ambos estados estables y una orbita que circunda a ambos puntos de equilibrio. Esta nueva orbita nos permite inferir que la energía que le entra al sistema es suficiente para oscilar con mayor amplitud que la amplitud de las oscilaciones iniciales, alrededor de otro punto de equilibrio.

```python
#Ahora definimos la función fourier que guarda en un array las frecuencias y el espectro de amplitudes de una señal
#Debemos incluir del paquete fftpack de scipy las funciones ifft, fft, fftfreq
from scipy.fftpack import ifft, fft, fftfreq
def fourier(window):
    h = window[1,0]-window[0,0]
    freq = fftfreq(len(window), h)
    absffty=np.abs(fft(window[:,1]))
    fdata = np.array((freq,absffty))
    return fdata

#Ahora aplicamos la función definida a cada una de las ventanas que se quieren analizar    
fwindow1 = fourier(window1)
fwindow2a = fourier(window2a)
fwindow2b = fourier(window2b)
fwindow3a = fourier(window3a)
fwindow3b = fourier(window3b)
fwindow4a = fourier(window4a)
fwindow4b = fourier(window4b)
fwindow4c = fourier(window4c)
fwindow5a = fourier(window5a)
fwindow5b = fourier(window5b)
fwindow5c = fourier(window5c)
```

```python
#Para mostrar los resultados del análisis de fourier, escogemos graficar el espectro de frecuencias de cada
#ventana para cada velocidad del viento. Los colores de los puntos del espectro coinciden con las lineas
#entrecortadas en la primera gráfica y con las orbitas del espacio de fases de la anterior gráfica.
fig = figure(figsize=(15,15))
fig.suptitle(u"DFT",fontsize=20)
ax1 = plt.subplot2grid((3,2),(0,0))
ax1.plot(fwindow1[0],fwindow1[1],"go",label = "ventana")
for i in range(len(fwindow1[0])):
    ax1.plot([fwindow1[0,i],fwindow1[0,i]],[0,fwindow1[1,i]],"b")
ax1.set_xlim(5,20)
ax1.set_ylim(200,2000)
ax1.set_title('Velocidad del viento = '+str(velocity[0])+' m/s')
ax1.grid()
ax1.legend()
ax2 = plt.subplot2grid((3,2),(0,1))
ax2.plot(fwindow2a[0],fwindow2a[1],"go",label = "ventana 1")
ax2.plot(fwindow2b[0],fwindow2b[1],"ro",label = "ventana 2")
for i in range(len(fwindow2a[0])):
    ax2.plot([fwindow2a[0,i],fwindow2a[0,i]],[0,fwindow2a[1,i]],"b")
for i in range(len(fwindow2b[0])):
    ax2.plot([fwindow2b[0,i],fwindow2b[0,i]],[0,fwindow2b[1,i]],"b")
ax2.set_xlim(5,20)
ax2.set_ylim(150,1500)
ax2.set_title('Velocidad del viento = '+str(velocity[1])+' m/s')
ax2.grid()
ax2.legend()
ax3 = plt.subplot2grid((3,2),(1,0))
ax3.plot(fwindow3a[0],fwindow3a[1],"go",label = "ventana 1")
ax3.plot(fwindow3b[0],fwindow3b[1],"ro",label = "ventana 2")
for i in range(len(fwindow3a[0])):
    ax3.plot([fwindow3a[0,i],fwindow3a[0,i]],[0,fwindow3a[1,i]],"b")
for i in range(len(fwindow3b[0])):
    ax3.plot([fwindow3b[0,i],fwindow3b[0,i]],[0,fwindow3b[1,i]],"b")
ax3.set_xlim(5,20)
ax3.set_ylim(220,4500)
ax3.set_title('Velocidad del viento = '+str(velocity[2])+' m/s')
ax3.grid()
ax3.legend()
ax4 = plt.subplot2grid((3,2),(1,1))
ax4.plot(fwindow4a[0],fwindow4a[1],"go",label = "ventana 1")
ax4.plot(fwindow4b[0],fwindow4b[1],"ro",label = "ventana 2")
ax4.plot(fwindow4c[0],fwindow4c[1],"co",label = "ventana 3")
for i in range(len(fwindow4a[0])):
    ax4.plot([fwindow4a[0,i],fwindow4a[0,i]],[0,fwindow4a[1,i]],"b")
for i in range(len(fwindow4b[0])):
    ax4.plot([fwindow4b[0,i],fwindow4b[0,i]],[0,fwindow4b[1,i]],"b")
for i in range(len(fwindow4c[0])):
    ax4.plot([fwindow4c[0,i],fwindow4c[0,i]],[0,fwindow4c[1,i]],"b")
ax4.set_xlim(1,25)
ax4.set_ylim(10,1600)
ax4.set_title('Velocidad del viento = '+str(velocity[3])+' m/s')
ax4.grid()
ax4.legend()
ax5 = plt.subplot2grid((3,2),(2,0),colspan=2)
ax5.plot(fwindow5a[0],fwindow5a[1],"go",label = "ventana 1")
ax5.plot(fwindow5b[0],fwindow5b[1],"ro",label = "ventana 2")
ax5.plot(fwindow5c[0],fwindow5c[1],"co",label = "ventana 3")
for i in range(len(fwindow5a[0])):
    ax5.plot([fwindow5a[0,i],fwindow5a[0,i]],[0,fwindow5a[1,i]],"b")
for i in range(len(fwindow5b[0])):
    ax5.plot([fwindow5b[0,i],fwindow5b[0,i]],[0,fwindow5b[1,i]],"b")
for i in range(len(fwindow5c[0])):
    ax5.plot([fwindow5c[0,i],fwindow5c[0,i]],[0,fwindow5c[1,i]],"b")
ax5.set_xlim(1,15)
ax5.set_ylim(40,4000)
ax5.set_title('Velocidad del viento = '+str(velocity[4])+' m/s')
ax5.grid()
ax5.legend()
fig.text(0.45,0.08,u'Frecuencia [Hz]',rotation='horizontal',fontsize=12)
plt.show()
```

Para una velocidad de 11.5 m/s, sólo se observa una frecuencia de resonancia y su siguiente armónico. Para las demas velocidades hay un ancho de banda de las frecuencias que pueden exitar la lámina biestable. En la siguiente tabla se resume lo observado.

Velocidad del viento (m/s)  | Tipo de Régimen | Frecuencias [Hz]
  ------------- | -------------| -------------
  11.5  | Lineal | 7.5
  12.2  | Caótico | 5 - 15
  13  | Caótico | 6 - 15 
  13.4  | Caótico | 1 - 24
  13.9  | Caótico | 1.4 - 15
  
Una vez se ha llegado al régimen caótico, el ancho de banda de las frecuencias de excitación aumentan al aumentar la velocidad del viento. Hasta que se llega a una velocidad de 13.9 m/s, donde el ancho se ha reducido, pues de acuerdo a lo observado, nos acercamos a velocidades que regresan el sistema al régimen lineal, caracterizado por tener una sola frecuencia de resonancia.

```python
#Por último se grafica los puntos de equilibrio en función de la velocidad del viento. Para hallarlos se encuentra
#el promedio de la posición angular en cada una de las ventanas con la función average de numpy. En las ventanas
#seleccionadas, la lámina oscila alrededor del punto de equilibrio esto quiere decir que el promedio de los datos
#de posición angular corresponde al punto de equilibrio.
fixed1 = np.average(window1[:,1])
fixed2a = np.average(window2a[:,1])
fixed2b = np.average(window2b[:,1])
fixed3a = np.average(window3a[:,1])
fixed3b = np.average(window3b[:,1])
fixed4a = np.average(window4a[:,1])
fixed4b = np.average(window4b[:,1])
fixed4c = np.average(window4c[:,1])
fixed5a = np.average(window5a[:,1])
fixed5b = np.average(window5b[:,1])
fixed5c = np.average(window5c[:,1])
#Se gráfica el diagrama de bifurcación usando herramientas de matplotlib: Se usa un ciclo que grafica
#los puntos de equilibrio en función de la velocidad. Para ello guardamos los resultados de
#los puntos de equilibrio en el array fixed.
fixed = np.array([fixed1,fixed2a,fixed2b,fixed3a,fixed3b,fixed4a,fixed4b,fixed4c,fixed5a,fixed5b,fixed5c])
fig = figure(figsize=(12,7))
fig.suptitle(u"Diagrama de Bifurcación",fontsize=20)
for i in range(5):
    if i == 0:
        plt.scatter(velocity[i],fixed[i])
    elif i == 1 or i == 2:
        plt.scatter(velocity[i],fixed[2*i-1])
        plt.scatter(velocity[i],fixed[2*i])
    elif i == 3:
        plt.scatter(velocity[i],fixed[2*i-1])
        plt.scatter(velocity[i],fixed[2*i])
        plt.scatter(velocity[i],fixed[2*i+1])
    else:
        plt.scatter(velocity[i],fixed[2*i])
        plt.scatter(velocity[i],fixed[2*i+1])
        plt.scatter(velocity[i],fixed[2*i+2])
        
plt.xlabel(u"Velocidad del viento [m/s]",size=12)
plt.ylabel(r'$\theta_{eq}$ [grados]',size=12)
plt.show()
```

Aqui podemos notar el mecanismo de duplicación del periodo, una vez se ha llegado a la velocidad del viento de 12.2 m/s. A partir de una velocidad de 13.4 m/s existen más puntos de equilibrio, para una sola velocidad del viento. Esta es otra evidencia del régimen caótico de las láminas a velocidades mayores a 12.2 m/s. A pesar de que sólo existe el registro de pocas velocidades del viento, podemos notar en este gráfico de bifurcación que se promueve el paso de un estado estable a otro con el perfil aerodinámico. Esto quiere decir que este perfil cumple su objetivo de excitar, a partir de las cargas aerodinámicas en el túnel de viento, el régimen caótico de las láminas biestables.
