## Introducción

Es de ínteres del ingeniero mecánico idear nuevas formas de reducir las pérdidas en sistemas de conversión, transporte y transformación de energía. Esto se logra produciendo dispositivos que aprovechen la energía residual del sistema, ya sea en forma de calor o de vibraciones mecánicas, y generen otra forma de energía. Esta área de la ingeniería es llamada harvesting. Para aprovechar las energía que brinda las vibraciones se ha diseñado un sistema que consiste en una lámina biestable y en un perfil aerodinámico.

###Lámina biestable
En el proceso de producción de la lámina, que es compuesta por un apilamiento de segmentos de fibra de carbono unidireccional, se generan esfuerzos residuales debido a la contracción y dilatación de la lámina en direcciones perpendiculares sobre el plano xy. Esto genera la biestabilidad de la lámina.

La biestabilidad brinda varias características a este sistema, sin embargo, la posibilidad de tener distintos comportamientos al estar excitada por un movimiento oscilatorio con cierta frecuencia es la característica por la que se usa en el área de harvesting. Cuando la lámina esta sujeta a un movimiento oscilatorio, existen 2 respuestas de la lámina biestable:


- Al estar excitada por una frecuencia igual a la frecuencia de resonancia de alguno de los estados estables, pasa de un estado estable a otro y se quede en el segundo estado, esto es llamado snap-through.

- Al estar excitada por distintas frecuencia determinadas por un rango, pase de un estado estable a otro aperiódicamente de forma caótica, esto es llamado régimen caótico.

###Perfil aerodinámico
En la universidad se encontró que la aceleración del movimiento armónico debido a las vibraciones de sistemas mecánicos no es suficiente para generar el régimen caótico en las láminas producidas, es por eso que se seleccionó un perfil aerodinámico. El perfil aerodinámico fue usado para idear otra forma de excitar el sistema a partir de cargas aerodinámicas. Se caracteriza por tener geometrias en su superficie superior e inferior que producen una diferencia de presión entre las superficies. Esta diferencia de presión da lugar a dos fuerzas importantes en el movimiento del perfil. La fuerza de sustentación es la fuerza vertical debido a la diferencia de presiones y la fuerza de arrastre es la fuerza horizontal debido a la diferencia de presiones. Además las fuerzas dependen del ángulo entre el flujo y el perfil (ángulo de ataque), del area del perfil y el largo del perfil. Es por esto que es común encontrar graficas del desempeño de un perfil en función del ángulo de ataque.

![Fuerza de sustentanción del perfil NACA 4418 en función del ángulo de ataque. Tomado de: Selig (1995).](Poner imagen del NACA)

###Lamina biestable como sistema no lineal
Un sistema no lineal se puede describir a partir de una matriz 2$\times$2, al expresar el movimiento con las variables de posición y velocidad, de la siguiente manera:

$A = \left( \begin{array}{ccc}
a & b \\
c & d \\ \end{array} \right)$

El polinomio característico del sistema se encuentra de calcular $det(A - \lambda I) = 0$, donde I es la matriz identidad. Las raíces resultantes de este polinomio son:

$\lambda_{1} = \frac {\tau+\sqrt{\tau^2-4\Delta}}{2}$
$\lambda_{2} = \frac {\tau-\sqrt{\tau^2-4\Delta}}{2}$

Donde $\Delta$ y $\tau$ son el determinante y la traza de la matriz $A$, respectivamente.

Sí las raíces del polinomio son reales, los puntos de equilibrio del sistema son llamados nodos, que pueden ser puntos silla, puntos estables o puntos inestables. Sí las raíces del polinomio son complejas, los puntos de equilibrio del sistema son llamados centros o espirales. Los centros son neutralmente estables, mientras que las espirales pueden ser estables o inestables. 

![Clasificación de los puntos de equilibrio en función del determinante y la traza de la matriz A. Tomado de: Strogatz (1994).](Poner imagen)

Adicionalmente, podemos observar el tipo de punto de equilibrio al graficar en el espacio de fases la variable de estado y su derivada.

![Espacio de fases de puntos de equilibrio típicos de sistemas no lineales. Tomado de: Strogatz (1994).](Poner imagen)

Las orbitas son características de sistemas que tienen soluciones periódicas, como lo son los osciladores armónicos simples que oscilan con cierta frecuencia. Mientras que las espirales representan osciladores no lineales que al tener amortiguamiento pueden tener oscilaciones que decaen o aumentan exponencialmente. Estas corresponden a espirales estables e inestables, respectivamente. Estos tipos de puntos de equilibrio son observables experimentalmente en sistemas modelados por la ecuación de Duffing.

Es por esto que se ha modelado el sistema a partir de un balance de fuerzas en donde la fuerza que deflecta la viga en voladizo, es decir la lámina biestable, se explica a partir de la ecuación del resorte de Duffing, que consiste en un sistema no lineal de masa-resorte:

$m\ddot{z} + c\dot{z} + k_1z + k_2z^2 + k_3z^3 = -m\ddot{y}$

Con la ayuda de la siguiente figura podemos entender esta ecuación. 

![Diagrama del modelo físico de la lámina biestable. Tomado de: Monsalve (2014)]()

La variable que designa la aceleración de la base es $\ddot{y}$ y la variable que designa la deflexión de la viga es $z$. Adicional al término lineal debido al resorte existen dos términos adicionales. El término cúbico de la ecuación, origina un potencial de doble pozo en el sistema de viga en voladizo. Evidencia experimental ha mostrado que existe un comportamiento asimétrico de la lámina biestable, en donde uno de los estados es más estable que el otro. Es por esto que se incluye un término cuadrático a la ecuación. Las constantes $m$, $c$, $k_1$, $k_2$ y $k_3$ dan cuenta de las propiedades de inercia, disipación de la energía y restauración de la energía del sistema biestable.

###Resultados y Análisis

```python
def derivate(data):
    h = data[1,0] - data[0,0]
    dataforward = np.roll(data[:,1],-1)
    databackward = np.roll(data[:,1],1)
    conv = np.pi/180
    omega = conv*(dataforward-databackward)/(2.*h)
    return omega[1:-1]
omega1 = derivate(data1)
omega2 = derivate(data2)
omega3 = derivate(data3)
omega4 = derivate(data4)
omega5 = derivate(data5)
tiempo = np.array([data1[1:-1,0],data2[1:-1,0],data3[1:-1,0],data4[1:-1,0],data5[1:-1,0]])
omega = np.array([omega1,omega2,omega3,omega4,omega5])
fig = figure(figsize=(15,15))
fig.suptitle(u"Velocidad angular en función del viento - Diferencias centrales",fontsize=20)
for i in range(5):
    if i == 0 or i == 1:
        ax = plt.subplot2grid((3,2),(0,i))
        ax.plot(tiempo[i],omega[i])
        ax.set_xlim(0,max(tiempo[i]))
        ax.set_title('Velocidad del viento = '+str(velocity[i])+' m/s')
    elif i == 2 or i == 3:
        ax = plt.subplot2grid((3,2),(1,i-2))
        ax.plot(tiempo[i],omega[i])
        ax.set_xlim(0,max(tiempo[i]))
        ax.set_title('Velocidad del viento = '+str(velocity[i])+' m/s')
    else:
        ax = plt.subplot2grid((3,2),(2,0),colspan=2)
        ax.plot(tiempo[i],omega[i])
        ax.set_xlim(0,max(tiempo[i]))
        ax.set_title('Velocidad del viento = '+str(velocity[i])+' m/s')
fig.text(0.08,0.55,r'$\omega$ [rad/s]',rotation='vertical', fontsize=12)
fig.text(0.5,0.08,u'Tiempo [s]',rotation='horizontal', fontsize=12)
plt.show()
```

```python
from scipy.fftpack import ifft, fft, fftfreq
def fourier(window):
    #fdata = []
    h = window[1,0]-window[0,0]
    freq = fftfreq(len(window), h)
    absffty=np.abs(fft(window[:,1]))
    fdata = np.array((freq,absffty))
    return fdata
    
fwindow1 = fourier(window1)
fwindow2a = fourier(window2a)
fwindow2b = fourier(window2b)
fwindow3a = fourier(window3a)
fwindow3b = fourier(window3b)
fwindow4a = fourier(window4a)
fwindow4b = fourier(window4b)
fwindow4c = fourier(window4c)
fwindow5a = fourier(window5a)
fwindow5b = fourier(window5b)
fwindow5c = fourier(window5c)
```
